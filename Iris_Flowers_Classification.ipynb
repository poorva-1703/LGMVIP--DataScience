Dataset:- http://archive.ics.uci.edu/ml/datasets/iris

```python
import numpy as np #array & Martices, numerical values
import matplotlib.pyplot as plt #grass
import seaborn as sns #statical graph
import pandas as pd  #Data Analysis 
 
```

```python
columns = ['Sepal length','Sepal width','petal length','Petal width','Class_labels'] #Load data
pf = pd.read_csv('iris.data', names=columns)
pf.head(10)            
```

```python
pf.describe()
```

```python
sns.pairplot(pf, hue= "Class_labels")
```

```python
data =pf.values

x =data[:,0:4]
y =data[:,4]
print(y)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train ,y_test = train_test_split(x, y, test_size=0.2)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

# ***Supervised Learning Algorithms***

**1.SVM (support vector machine)**

```python
 from sklearn.svm import SVC

model_svc =SVC()
model_svc.fit(x_train, y_train)
```

```python
prediction1 = model_svc.predict(x_test)
#Accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1)*100)
for i in range (len(prediction1)):
  print(y_test[i],prediction1[i])
```

**2.Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
prediction2 = model_LR.predict(x_test) #accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100) 
for i in range (len(prediction2)):
  print(y_test[i],prediction2[i])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction2))
```

```python
x_new = np.array ([[3,2,1,0.2],[4.9,2.2,3.8,1.1],[5.3,2.5,4.6,1.9]])

prediction = model_svc.predict(x_new)
print('prediction of species: {}'.format(prediction))
```
